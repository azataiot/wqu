```python
from wqu.dp import GBM
import matplotlib.pyplot as plt

gbm = GBM(S0=100, mu=0.05, sigma=0.2, T=1, N=252)
```

```python
gbm.simulate()
```

```python
gbm.simulate(method="euler")
```

```python
gbm.plot(M=1, method='exact')
```

```python
gbm = GBM(S0=100, mu=0.05, sigma=0.2)
prices = gbm.simulate(method="exact")
print(prices[:5])
```

```python
# Create a GBM simulator
gbm = GBM(S0=100, mu=0.05, sigma=0.2, T=1, N=252, seed=42)
```

```python
# Simulate one exact path and print first 5 prices
path = gbm.simulate(method="exact")
print(path[:5])
```

```python
# Plot a single Euler path
gbm.plot(M=1, method="euler", color='green', lw=2)
```
